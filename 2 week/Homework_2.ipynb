```python
n = len(new_big_data)
n_test = n_val = int(n * 0.2)
n_train = n - n_val - n_test
n, n_train + n_val + n_test, n_test, n_val, n_train
```

```python
idx = np.arange(n)
np.random.seed(9)
np.random.shuffle(idx)
```

```python
df_train = new_big_data.iloc[idx[:n_train]]
df_val = new_big_data.iloc[idx[n_train:n_train + n_val]]
df_test = new_big_data.iloc[idx[n_train+n_val:]]

df_train
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

df_train.head(10)
```

```python
y_train = np.log1p(df_train["median_house_value"].values)
y_val = np.log1p(df_val["median_house_value"].values)
y_test = np.log1p(df_test["median_house_value"].values)

y_train
```

```python
del df_train["median_house_value"]
del df_val["median_house_value"]
del df_test["median_house_value"]

df_train
```

```python
def prepare_zero_X(df):
    df_train_with_zeros = df.copy()
    df_train_with_zeros['total_bedrooms'] = df_train_with_zeros['total_bedrooms'].fillna(0)
    return df_train_with_zeros.values
```

```python
def prepare_mean_X(df):
    df_train_with_mean = df.copy()
    mean_value = df_train['total_bedrooms'].mean()
    df_train_with_mean['total_bedrooms'] = df_train_with_mean['total_bedrooms'].fillna(mean_value)
    return df_train_with_mean.values
```

```python
def train_linear_regression(X, y):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])
    
    XTX = X.T.dot(X)
    XTX_inv = np.linalg.inv(XTX)
    w_full = XTX_inv.dot(X.T).dot(y)
    
    return (w_full[0], w_full[1:])

def rmse(y, y_pred):
    se = (y - y_pred) ** 2
    mse = se.mean()
    return np.sqrt(mse)
```

```python
X_train_zeros = prepare_zero_X(df_train)
w0_0, w_0 = train_lnear_regression(X_train_zeros, y_train)

X_val = prepare_zero_X(df_val)
y_pred_0 = w0_0 + X_val.dot(w_0)

rmse(y_val, y_pred_0)
```

```python
X_train_means = prepare_mean_X(df_train)
w0_1, w_1 = train_lnear_regression(X_train_means, y_train)

X_val = prepare_mean_X(df_val)
y_pred_1 = w0_1 + X_val.dot(w_1)
rmse(y_val, y_pred_1)
```

```python
def train_lnear_regression_reg(X, y, r):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])
    
    XTX = X.T.dot(X)
    XTX = XTX + r * np.eye(XTX.shape[0])
    XTX_inv = np.linalg.inv(XTX)
    w_full = XTX_inv.dot(X.T).dot(y)
    
    return (w_full[0], w_full[1:])
```

```python
rs = [0, 0.000001, 0.0001, 0.001, 0.01, 0.1, 1, 5, 10]

for r in rs:
    X_train_zeros = prepare_zero_X(df_train)
    w0_0, w_0 = train_lnear_regression_reg(X_train_zeros, y_train, r)

    X_val = prepare_zero_X(df_val)
    y_pred_0 = w0_0 + X_val.dot(w_0)

    print(rmse(y_val, y_pred_0))
```
